```python
postId = []
urls = []
for idx in data.index:
    id = data.loc[idx, 'postId']
    photos = data.loc[idx, 'photos']
    for photo in photos:
        postId.append(id)
        urls.append(photo)

data_urls = pandas.DataFrame({'postId': postId, 'urls': urls})
```
